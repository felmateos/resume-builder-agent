```python
from jinja2 import Environment, FileSystemLoader
from pathlib import Path
import json
```

```python
env = Environment(
    loader=FileSystemLoader("04_templates"),
    block_start_string="((*",
    block_end_string="*))",
    variable_start_string="(((",
    variable_end_string=")))",
    comment_start_string="((#",
    comment_end_string="#))",
    autoescape=False
)
```

```python
with open("./03_jsons/gap_analysis.json", "r", encoding="utf-8") as f:
    context = json.load(f)
```

```python
template = env.get_template('gaps.tex')
output = template.render(**context)
```

```python
output_path = Path("05_output/gaps.tex")

with output_path.open("w", encoding="utf-8") as f:
    f.write(output)
```
