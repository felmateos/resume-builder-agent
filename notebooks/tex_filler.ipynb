```python
from jinja2 import Environment, FileSystemLoader
```

```python
context = {
    "gaps": [
        {
            "title": "Falta de experiência com Spark",
            "description": "A vaga exige uso intensivo de Spark em produção.",
            "actions": [
                "Refazer pipeline pessoal usando Spark",
                "Estudar Spark SQL e otimizações",
                "Implementar job com dados reais"
            ]
        }
    ]
}
```

```python
def render_gaps_latex(gaps_data):
    env = Environment(loader=FileSystemLoader("templates"))
    template = env.get_template("gaps.tex")
    return template.render(**gaps_data)
```

```python
print(render_gaps_latex(context))
```
